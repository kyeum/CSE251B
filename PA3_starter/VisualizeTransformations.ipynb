```python
def vis_orgin(test_loader, title="Original Image"):
    inputimg = []
    for i, (img, label) in enumerate(test_loader):
        inputimg = img[0]
#             break # should be a break here, but alas we already used this everywhere else

    plt.axis('off')
    plt.imshow(inputimg)
    plt.title(title)
#     plt.savefig(f"DataAug_{title}")
    plt.show()
    
```

```python
# Original Image
test_dataset = TASDataset('tas500v1.1', eval_mode=True, mode='test', transform_mode=0)
# Should be setting batch_size=1, but since our visualize function gets the 1st image of the last batch, we need to 
# set to the same batch_size to get the same image.
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
vis_orgin(test_loader)
```

```python
# Horizontal Flip
test_dataset = TASDataset('tas500v1.1', eval_mode=True, mode='test', transform_mode=1)
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
vis_orgin(test_loader, "Horizontal Flip")
```
